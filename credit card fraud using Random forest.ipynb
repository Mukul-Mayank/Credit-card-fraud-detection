Data Visualisation,Data Engineering and handling outliers

```python
# Counting the number of columns
num_columns = data.shape[1]

print(f'The number of columns in the dataset is: {num_columns}')
```

```python
data.columns
```

```python
# List of columns to visualize and handle outliers
columns_to_check = data.columns.difference(['Time', 'Class'])
```

```python

# Visualizing the data before handling outliers
plt.figure(figsize=(20, 15))
for i, column in enumerate(columns_to_check, 1):
    plt.subplot(6, 5, i)
    sns.boxplot(data[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

```python

# Handling outliers using the IQR method
Q1 = data[columns_to_check].quantile(0.25)
Q3 = data[columns_to_check].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Removing outliers from all numeric features
for column in columns_to_check:
    lower_bound = Q1[column] - 1.5 * IQR[column]
    upper_bound = Q3[column] + 1.5 * IQR[column]
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
```

```python
# Visualizing the data after handling outliers
plt.figure(figsize=(20, 15))
for i, column in enumerate(columns_to_check, 1):
    plt.subplot(6, 5, i)
    sns.boxplot(data[column])
    plt.title(f'Boxplot of {column} (cleaned)')
plt.tight_layout()
plt.show()
```

Model Selection and Model building

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
from imblearn.over_sampling import SMOTE
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns=['Class'], axis=1)
y = data['Class']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalizing the 'Amount' column
scaler = MinMaxScaler()
data['Amount'] = scaler.fit_transform(data[['Amount']])
```

```python
data.hist(figsize = (20,20))
```

```python
sns.pairplot(data=data)
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 5))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False, cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Check for outliers in the 'Amount' column
plt.figure(figsize=(10, 3))
sns.boxplot(x=data['Amount']) #palette="icefire"
plt.title("Boxplot of Transaction Amount")
plt.xlabel("Transaction Amount")
plt.show()
```

```python
# Handle Class Imbalance with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

Train_test split

```python
#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

Model Building Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)
```

```python
rf_model.score(X_train,y_train)*100,rf_model.score(X_test,y_test)*100
```

Prediction and Evaluation on the trained model

```python
# Predictions
y_pred = rf_model.predict(X_test)
y_pred_prob = rf_model.predict_proba(X_test)[:, 1]
```

```python
# Evaluation
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
roc_auc = roc_auc_score(y_test, y_pred_prob)
print(f"ROC-AUC Score: {roc_auc:.2f}")
```

```python

# Visualising Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="Blues", xticklabels=["Legit", "Fraud"], yticklabels=["Legit", "Fraud"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
